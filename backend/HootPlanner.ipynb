USE **all-MiniLM-L6-v2** sentence embedder, specifically
mapping sentences & paragraphs to 384-dim dense vector space for clustering or semantic search tasks. FAST SPEED

Comparison:

```

all-MiniLM-L12-v2
Layers: 12 transformer layers.
Embedding Size: 384 dimensions.
Performance: Higher accuracy compared to the L6 version due to its deeper architecture. It tends to capture more detailed semantic information.
Inference Speed: Slower than the L6 version because of the larger model and more layers.
Model Size: 33 MB.
Use Case: If you prioritize better performance in terms of accuracy and can trade off some speed, this is a better option.


all-MiniLM-L6-v2
Layers: 6 transformer layers (half of L12).
Embedding Size: 384 dimensions (same as L12).
Performance: Slightly less accurate than the L12 version, but still performs well for many tasks. It may not capture as much nuance as the L12 version.
Inference Speed: Faster due to fewer layers, making it suitable for large-scale tasks where speed is critical.
Model Size: 22 MB.
Use Case: If you need faster inference and can accept slightly reduced accuracy, this is the better option.
```



````python
from transformers import AutoTokenizer, AutoModel
import torch

tokenizer = AutoTokenizer.from_pretrained("sentence-transformers/all-MiniLM-L12-v2")
model = AutoModel.from_pretrained("sentence-transformers/all-MiniLM-L12-v2")

# Function to generate embeddings
def generate_embeddings(text):
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True)
    outputs = model(**inputs)
    # embeddings = outputs.last_hidden_state.mean(dim=1)
    # embeddings = embeddings.detach().numpy().flatten()  # Flatten the embeddings
    # return embeddings
    return outputs.last_hidden_state.mean(dim=1).detach().numpy()
````

````python
test_text = "This is a test."
test_embeddings = generate_embeddings(test_text)

# check the shape
print("Embedding shape:", test_embeddings.shape)
````

````python

research_data['lab_description'].fillna('', inplace=True)
research_data['research_description'].fillna('', inplace=True)

combined_descriptions = (research_data['lab_description'] + " " + research_data['research_description']).tolist()

embeddings = generate_embeddings(combined_descriptions)

arr = [{
    "vector": embedding.flatten().tolist(),
    "faculty_member": research_data.iloc[i]['faculty_member'],
    "description": combined_descriptions[i]
} for i, embedding in enumerate(embeddings)]


# # Generate embeddings for all descriptions at once
# embeddings = generate_embeddings(descriptions)

# # Flatten and create the array
# arr = [{
#     "vector": embedding.flatten().tolist(),
#     "faculty_member": research_data.iloc[i]['faculty_member'],
#     "description": descriptions[i]
# } for i, embedding in enumerate(embeddings)]
````
